```python
import pandas as pd

# Carregar o arquivo CSV com a codificação correta
base = pd.read_csv('trab - fatos.csv.csv', encoding='ISO-8859-1')  # ou 'utf-8' se necessário

# Substituir todos os valores "MTE" por 1 no DataFrame
base = base.apply(lambda col: col.map(lambda x: 1 if str(x).strip().lower() == 'mte' else x) if col.dtypes == 'object' else col)

# Substituir todos os valores "2023" por 1 no DataFrame
base = base.applymap(lambda x: 1 if str(x).strip() == '2023' else x)

# Salvar o DataFrame modificado em um novo arquivo CSV
base.to_csv('trab - fatos_modificado.csv', index=False, encoding='ISO-8859-1')  # ou 'utf-8' para salvar
```

```python
import pandas as pd

# Carregar a primeira tabela
tabela1 = pd.read_csv('dim_servidor.csv', encoding='ISO-8859-1')  # Ajuste o caminho do arquivo

# Carregar a segunda tabela
tabela2 = pd.read_csv('trab - fatos_modificado.csv', encoding='ISO-8859-1')  # Ajuste o caminho do arquivo

# Garantir que o nome das colunas da segunda tabela estão corretos
# Se necessário, ajustar os nomes das colunas para a codificação correta
tabela2.columns = tabela2.columns.str.replace('�', '')  # Substitui caracteres errados

# Realizar a junção das duas tabelas com base no CPF
tabela_completa = pd.merge(tabela2, tabela1[['CPF', 'id_servidor']], on='cpf', how='left')

# Substituir o CPF na tabela completa pelo 'id_servidor'
tabela_completa['cpf'] = tabela_completa['id_servidor']

# Remover a coluna 'id_servidor' da tabela resultante, já que a substituição foi feita
tabela_completa = tabela_completa.drop(columns=['id_servidor'])

# Salvar o DataFrame resultante em um novo arquivo CSV
tabela_completa.to_csv('tabela_comparada.csv', index=False, encoding='ISO-8859-1')

```

```python
import pandas as pd
df_servidores = pd.read_csv('dim_servidor.csv')
df_remuneracoes = pd.read_csv('trab - fatos_modificado.csv')
# Suponha que 'df_servidores' seja a Tabela 1 e 'df_remuneracoes' seja a Tabela 2

# Criar um dicionário para mapear CPF para o ID do servidor da Tabela 1
cpf_para_id = dict(zip(df_servidores['cpf'], df_servidores['id_servidor']))

# Substituir a coluna cpf_2 da Tabela 2 pelo ID do servidor
df_remuneracoes['cpf_2'] = df_remuneracoes['cpf_2'].map(cpf_para_id)

# Ver o resultado
print(df_remuneracoes)
```

```python
import pandas as pd

# Carregar a Tabela 1
tabela_1 = pd.read_csv("dim_servidor.csv", delimiter=',', encoding='latin1')

# Carregar a Tabela 2
tabela_2 = pd.read_csv("trab - fatos.csv (1).csv", delimiter=',', encoding='latin1')

# Remover espaços em branco nas colunas, caso existam
tabela_1.columns = tabela_1.columns.str.strip()
tabela_2.columns = tabela_2.columns.str.strip()

# Criar um dicionário de mapeamento entre CPF e id_servidor
cpf_para_id = pd.Series(tabela_1['id_servidor'].values, index=tabela_1['CPF'].values).to_dict()

# Substituir os CPFs da Tabela 2 pelos id_servidor correspondentes
tabela_2['CPF'] = tabela_2['CPF'].map(cpf_para_id)

# Salvar a Tabela 2 modificada em um novo arquivo CSV
tabela_2.to_csv("trab - fatos_modificado.csv", index=False, encoding='latin1')

# Verificar se a modificação foi salva corretamente
print("Tabela 2 salva com sucesso em 'tabela2_modificada.csv'")
```

```python
import pandas as pd

# Carregar a Tabela 2
tabela_2 = pd.read_csv("trab - fatos_modificado.csv", delimiter=',', encoding='latin1')

# Remover espaços em branco nas colunas, caso existam
tabela_2.columns = tabela_2.columns.str.strip()

# Remover os ".0" da coluna 'CPF' (convertendo para inteiro)
tabela_2['CPF'] = tabela_2['CPF'].astype('Int64')  # 'Int64' para garantir que valores nulos (NaN) sejam mantidos

# Salvar a Tabela 2 modificada em um novo arquivo CSV
tabela_2.to_csv(".csv", index=False, encoding='latin1')

# Verificar se a modificação foi salva corretamente
print("Tabela 2 salva com sucesso em 'A.csv'")
```
